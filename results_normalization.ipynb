```python
import anndata as ad
import pandas as pd
from src.utils import get_model_stats
import pickle as pkl
```

## Raw Counts

```python
adata_ctrl_rawcounts = ad.read_h5ad("./data/normalization/sciplex_raw_filt.h5ad")
adata_ctrl_rawcounts = adata_ctrl_rawcounts[adata_ctrl_rawcounts.obs.product_name == "Vehicle"]
```

```python
with open("./results/null_rawcount_norm_res.pkl", 'rb') as f:
    df_raw_null = pkl.load(f)

with open("./results/mean_rawcount_norm_res.pkl", 'rb') as f:
    df_raw_mean = pkl.load(f)

with open("./results/decoder_rawcount_norm_res.pkl", 'rb') as f:
    df_raw_decoder = pkl.load(f)

with open("./results/mlp_rawcount_norm_res.pkl", 'rb') as f:
    df_raw_mlp = pkl.load(f)

with open("./results/prnet_rawcount_norm_res.pkl", 'rb') as f:
    df_raw_prnet = pkl.load(f)
```

```python
stats_raw_null = get_model_stats(df_raw_null,
                                 adata_ctrl_rawcounts, 
                                 list(adata_ctrl_rawcounts.var_names),
                                 'null_rawcount',
                                 dose_subset=1000, 
                                method_deg='wilcoxon')
```

```python
stats_raw_mean = get_model_stats(df_raw_mean,
                                 adata_ctrl_rawcounts, 
                                 list(adata_ctrl_rawcounts.var_names),
                                 'mean_rawcount',
                                 dose_subset=1000, 
                                method_deg='wilcoxon')
```

```python
stats_raw_decoder = get_model_stats(df_raw_decoder,
                                 adata_ctrl_rawcounts, 
                                 list(adata_ctrl_rawcounts.var_names),
                                 'decoder_rawcount',
                                 dose_subset=1000, 
                                method_deg='wilcoxon')
```

```python
stats_raw_mlp = get_model_stats(df_raw_mlp,
                                 adata_ctrl_rawcounts, 
                                 list(adata_ctrl_rawcounts.var_names),
                                 'mlp_rawcount',
                                 dose_subset=1000, 
                                method_deg='wilcoxon')
```

```python
stats_raw_prnet = get_model_stats(df_raw_prnet,
                                 adata_ctrl_rawcounts, 
                                 list(adata_ctrl_rawcounts.var_names),
                                 'prnet_rawcount',
                                 dose_subset=1000, 
                                method_deg='wilcoxon')
```
