```python
import pickle as pkl
import anndata as ad
import pandas as pd
from tqdm import tqdm

from src.evaluation.calculate_metrics import get_model_stats
```

```python
result_list = pd.read_csv("./experiment_list.txt", delimiter="\t")
```

```python
#sanity check
for i, row in result_list.iterrows():
    print(row)
    adata_control = ad.read_h5ad(row['adata_path'])
    gene_names = list(adata_control.var_names)

    with open(row['results_path'], 'rb') as f:
        df = pkl.load(f)

    del df
    del adata_control
```

```python
all_stats = list()

for i, row in tqdm(result_list.iterrows()):
    adata_control = ad.read_h5ad(row['adata_path'])
    gene_names = list(adata_control.var_names)

    with open(row['results_path'], 'rb') as f:
        df = pkl.load(f)

    model_name = row['model_name']
    experiment_key = row['experiment_key']
    experiment_condition = row['experiment_condition']
    if pd.isna(row['dose_subset']):
        dose_subset = None
    else:
        dose_subset = row['dose_subset']

    res = get_model_stats(test_results=df,
                         adata_control=adata_control,
                         gene_names=gene_names,
                         model_name=model_name,
                         experiment_key=experiment_key,
                         experiment_condition=experiment_condition,
                         dose_subset=dose_subset)

    all_stats.append(res)

    del df
    del adata_control

res_final = pd.DataFrame(all_stats)
res_final.to_pickle("results_final.pkl")
```
